### Data Visualisation

```python
df = pd.DataFrame(data)

plt.scatter(df['Average Meal Price'], df['Revenue'], color='grey', s= 1)
plt.title('Average Meal Price vs Revenue')
plt.xlabel('Average Meal Price ($)')
plt.ylabel('Revenue ($ million)')

plt.show()
```

We can see that as the average meal price increases the Revenue also increases
There's a positive correleation between Avereage meal price and Revenue.

When the lowest average meal price is 25, it results in revenue of 0.2 million
When the highest average meal price is 76, it results in revenue of 1.5 million

```python
df = pd.DataFrame(data)

plt.scatter(df['Seating Capacity'], df['Revenue'], color='brown', s=1 )
plt.title('Seating Capacity vs Revenue')
plt.xlabel('Seating Capacity')
plt.ylabel('Revenue ($ million)')

plt.show()
```

As the seating capacity increases, the revenue also increases. 
There's a positive correlation between 2 variables. 

At the lowest seating capacity of 30, the revenue is 0. 18 million
At the highest seating capacity of 90, the revenue is 1.5 million
This might suggest that more seats lead to more customers and higher sales.


```python

df = pd.DataFrame(data)

plt.scatter(df['Revenue'], df['Marketing Budget'], color='lightgreen', s=1)
plt.title('Marketing Budget vs Revenue')
plt.xlabel('Revenue ($ million)')
plt.ylabel('Marketing Budget ($)')

plt.show()
```

There is a bit of cluster at the bottom left hand corner but it is still showing that there is a positive correlation between Marketing budget and Revenue.
At the lowest marketing budget of 604, gives revenue of 0.18 million. 
At the highest marketing budget of 9978, gives revenue of 1.5 million. 
The cluster indicates that lower marketing budgets are associated with lower revenue, suggesting these are potentially less effective marketing efforts or newer initiatives. 
But the positive correlation indicates as the marketing budget increases, revenue tends to increase, indicating that higher investments in marketing generally lead to better revenue outcomes.

```python
df = pd.DataFrame(data)

plt.scatter(df['Social Media Followers'], df['Marketing Budget'], color='purple', s=0.5)
plt.title('Social Media Followers vs Marketing Budget')
plt.xlabel('Social Media Followers')
plt.ylabel('Marketing Budget ($)')
plt.show()
```

There is a correlation coefficient of 1, which indicates there's a strong positive linear relationship between social media followers and marketing budgets. 
Since increasing social media followers are associated with higher marketing budget, this suggest that businesses see value in investing more in social media as their follower base grows. This investment can lead to increased brand awareness, customer engagement, and revenue growth.   

```python
# to see how revenue differs by location.
data['Location'].value_counts()
```

```python
df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
plt.barh(df['Location'], df['Revenue'], color=['skyblue'])
plt.xlabel('Revenue ($ million)')
plt.ylabel('Location')
plt.title('Revenue by Location')
plt.xlim(0, df['Revenue'].max() + 500)

plt.show()
```

It is showing that the restaurants located in the downtown areas are showing highest Revenue of around 1.5 million compared to rural and suburban areas.

Whereas restaurants located in the rural areas are showing lowest average revenue around 0.9 million.

```python
df = pd.DataFrame(data)

# Create scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, 
                x='Average Meal Price', 
                y='Revenue', 
                hue='Location', 
                size='Seating Capacity', 
                sizes=(20, 300), 
                alpha=0.7, 
                palette='deep', 
                edgecolor='pink')


plt.title('Restaurant Revenue vs. Average Meal Price')
plt.xlabel('Average Meal Price')
plt.ylabel('Revenue')
plt.legend(title='Location', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

```python
df = pd.DataFrame(data)
numeric_columns= df.select_dtypes(include=['float64', 'int64']).columns
corr_matrix= df[numeric_columns].corr()

plt.figure(figsize=(12,8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt= '.2f', vmin=-1, vmax=1)
plt.title('Correlation Heatmap')
plt.show()
```

From this we can see that Revenue depends mainly on Seating capacity, Average meal price, 
and Marketing budget.

As we can see from the chart, all of the features show positive correlation with Revenue.
